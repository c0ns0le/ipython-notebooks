```python
import requests
import lxml.html
from lxml.html.clean import Cleaner 

site_url = "http://free-tv-video-online.me/internet/"

html = requests.get(site_url).content
cleaner = Cleaner(page_structure=False)
cleaned_html = cleaner.clean_html(html)
root = lxml.html.fromstring(cleaned_html)

for tbl in root.xpath('//table[@cellspacing="1"]'):
    elements = tbl.xpath('.//tr//td//b//text()')
    links = [el.get('href') for el in tbl.xpath('.//tr//td//a')]
    print elements
    print links
    print len(elements), len(links)
```

```python
site_url = "http://www.free-tv-video-online.me/internet/modern_family/season_6.html"

html = requests.get(site_url).content
cleaner = Cleaner(page_structure=False)
cleaned_html = cleaner.clean_html(html)
root = lxml.html.fromstring(cleaned_html)

for tbl in root.xpath('//table[@cellpadding="4"]'):
    elements = tbl.xpath('.//tr//td//b//text()')
    links = [(el.get('href'), el.xpath('.//div//text()')) for el in tbl.xpath('.//tr//td//a') if el.get('href') and not any(text in el.get('href') for text in ['report', 'direct_connect'])]
    print elements
    print links
    print len(elements), len(links)
```
