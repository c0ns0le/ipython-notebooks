```python
import requests
```

You can create different filters to specify the return value of the json object
by default the returned question or answer object doesn't have a body attribute.

```python
r = requests.get('https://api.stackexchange.com/2.2/search/advanced?order=desc&sort=votes&tagged=javascript&wiki=True&site=stackoverflow')
```

```python
len(r.json()['items'])
```

```python
r.json()['items'][0]
```

```python
r_answers = requests.get('https://api.stackexchange.com/2.2/questions/111102/answers?order=desc&sort=votes&site=stackoverflow')
```

```python
r_answers.json()
```

```python
len(r_answers.json()['items'])
```

```python
r_answer = requests.get('https://api.stackexchange.com/2.2/answers/111200?order=desc&sort=activity&site=stackoverflow')
```

```python
r_answer.json()
```
