```python
from collections import deque, defaultdict

def solve(N, d):

    char_to_num = {int(item, 16): item  for item in d}
    d = sorted(char_to_num.keys())
    prev = defaultdict(lambda: -2)
    digit = defaultdict(int)

    q = deque()
    for i in xrange(len(d)):
        if d[i] !=0 and prev[d[i]%N] == -2:
            prev[d[i]%N] = -1
            digit[d[i]%N] = d[i]
            q.append(d[i]%N)

    while q and prev[0] == -2:
        curr = q.popleft()
        for i in xrange(len(d)):
            next_number = ((curr*16)%N + d[i])%N
            if prev[next_number] == -2:
                prev[next_number] = curr
                digit[next_number] = d[i]
                q.append(next_number)
    if prev[0] == -2:
        print "no solution"
    else:
        s = char_to_num[digit[0]]
        t = prev[0]
        while t != -1:
            s = char_to_num[digit[t]] + s
            t = prev[t]
        print s
```
